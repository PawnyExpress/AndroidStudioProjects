**Hole in a Graph**

$$ y = \frac{3(x-2)}{x-2} $$

```python
x = 2
h = 0.00001

y_right = (3*((x+h)-2))/(x+h-2)
y_left = (3*((x-h)-2))/((x-h)-2)

print("y_right = ", y_right)
print("y_left = ", y_left)

if round(y_right) != round(y_left):
  print("Limit does not exist at x = ", x)
```

**Limits - where is the function undefined?**

$$ y = \frac{3x^2}{x^2-4} $$

```python
import sympy
from sympy import symbols
from sympy.solvers import solve

x = symbols('x')

#Place equation here
eq = x**2-4

#solve() sets the equation equal to zero
print("x = ", solve(eq, x))
```

**Limits - approaching from the right or the left.**
$$ y = \frac{3x^2}{x^2-4} $$

```python
x = 2
h= 0.0001

# For y = (3*x**2)/(x**2 - 4)

y_right = (3*(x+h)**2)/((x+h)**2 - 4)
y_left = (3*(x-h)**2)/((x-h)**2 - 4)

print("y_right = ", y_right)
print("y_left = ", y_left)

if round(y_right) != round(y_left):
  print("Limit does not exist at x = ", x)
```

**Limits - as x approaches infinity**
$$ y = \frac{3x^2}{x^2-4} $$

```python
# For equation y = (3*x**2)/(x**2 - 4)

for x_value in [10, 100, 1000, 10000]:
  limit = (3*x_value**2)/(x_value**2 - 4)
  print("x = ", x_value, " Limit = ", limit)
```

**Limits - graphing**

Example function: $ y = \frac{3x^2}{x^2-4} $

```python
import matplotlib.pyplot as plt
import numpy as np

# for equation y = (3*x**2)/(x**2 - 4)
x_value = 0
y_value = (3*x_value**2)/(x_value**2 - 4)

zoom = 10
xmin = x_value - zoom
xmax = x_value + zoom
ymin = y_value - zoom
ymax = y_value + zoom

x = np.linspace(xmin, xmax, 100)
y = (3*x**2)/(x**2 - 4)
#y = (3*x*(x-2)/(x-2)

plt.axis([xmin, xmax, ymin, ymax]) #window size
plt.plot(x,y)
plt.plot([x_value], [y_value], 'ro')
#plt.plot([xmin,xmax], [3,3], 'r--')
plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')
plt.xlabel("x")
plt.ylabel("y")
plt.title('Graph of y = (3*x**2)/(x**2 - 4)')
plt.show()
```

**Limits with Sympy**

Example function: $ y = \frac{3x^2}{x^2-4} $

```python
import sympy
from sympy import symbols

x = symbols('x')

# Put the eqwuation here
eq = (3*x**2)/(x**2 - 4)

# critical values
x_1 = 2
x_2 = -2

# limit(expression, variable, value)

# first critical point
right = sympy.limit(eq, x, x_1, dir='+')
left = sympy.limit(eq, x, x_1, dir='-')
print("at x = ", x_1)
print("right = ", right)
print("left = ", left)

# second critical point
right = sympy.limit(eq, x, x_2, dir='+')
left = sympy.limit(eq, x, x_2, dir='-')
print("at x = ", x_2)
print("right = ", right)
print("left = ", left)

#infinity
right = sympy.limit(eq, x, sympy.oo)
left = sympy.limit(eq, x, -sympy.oo)
print("as x approaches infinity: ", right)
print("as x approaches negative infinity: ", left)
```
